```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Replace Unknowns

```python
data.replace("unknown",np.nan,inplace=True)
```

```python
data['job'].unique()
```

```python
data.isna().sum()
```

```python
data.drop('poutcome',axis=1,inplace=True)
```

```python
data['education'].replace(np.nan,data['education'].mode()[0],inplace=True)
```

# One Hot Encoding

```python
from sklearn import preprocessing
enc = preprocessing.OneHotEncoder()
x = enc.fit_transform(data[['job']].values).toarray()
x
```

```python
df = pd.DataFrame(data=x,columns = ["job"+str(int(i)) for i in range(x.shape[1])])
df = df.iloc[:,:-1]
df
```

```python
dataset = pd.concat([data,df],axis=1)
dataset
```

```python
dataset.drop('job',axis=1,inplace=True)
```

```python
x = enc.fit_transform(data[['job']].values).toarray()
```

```python
df = pd.DataFrame(data=x,columns = ["marital"+str(int(i)) for i in range(x.shape[1])])
df = df.iloc[:,:-1]
df
```

```python
data = pd.concat([dataset,df],axis=1)
```

```python
data.drop('marital',axis=1,inplace=True)
```

```python
x = enc.fit_transform(data[['education']].values).toarray()
df = pd.DataFrame(data=x,columns = ["education"+str(int(i)) for i in range(x.shape[1])])
df = df.iloc[:,:-1]
df
```

```python
dataset = pd.concat([data,df],axis=1)
dataset
```

```python
dataset.drop('education',axis=1,inplace=True)
```

```python
x = enc.fit_transform(data[['month']].values).toarray()
df = pd.DataFrame(data=x,columns = ["month"+str(int(i)) for i in range(x.shape[1])])
df = df.iloc[:,:-1]
df
```

```python
data = pd.concat([dataset,df],axis=1)
data.drop('month',axis=1,inplace=True)
data
```

```python
data.drop('duration',axis=1,inplace=True)
```

# Get Dummies

```python
data = pd.get_dummies(data=data,columns=['default','housing','loan','contact'],drop_first=True)
```

```python
data
```

## Label Encoding for Target

```python
from sklearn import preprocessing
enc = preprocessing.LabelEncoder()
data['Y'] = enc.fit_transform(data['y'])
```

```python
data.drop('y',axis=1,inplace=True)
data
```

```python
sizes = data['Y'].value_counts(sort = True)
colors = ["yellow","green"] 
plt.pie(sizes, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90,)
plt.title('Percentage of target in the dataset')
plt.show()
```

```python
X = data.drop('Y', axis=1)
Y = data['Y']
from sklearn.datasets import make_classification
from imblearn.under_sampling import RandomUnderSampler
X, Y = make_classification(n_samples=10000, weights=[0.99], flip_y=0)
undersample = RandomUnderSampler(sampling_strategy='majority')
x, y = undersample.fit_resample(X,Y)
```

# Split

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y)
```

```python
model.fit(train_x,train_y)
```
